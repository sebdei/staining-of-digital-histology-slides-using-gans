```python
from matplotlib import pyplot as plt
import cv2
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import colors


path = "10-14-94F1DA-B-E3C7B2-B.ihc.HB.jpg"

img = cv2.imread(path)
img = cv2.resize(img, (512, 512))

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title('my picture')
plt.show()
```

```python
pixel_colors = img.reshape((np.shape(img)[0]*np.shape(img)[1], 3))
norm = colors.Normalize(vmin=-1.,vmax=1.)
norm.autoscale(pixel_colors)
pixel_colors = norm(pixel_colors).tolist()
```

```python
fig = plt.figure(figsize=(12,8), dpi= 100)
ax = fig.add_subplot(111, projection='3d')

ax.view_init(30, 45)

h, s, v = cv2.split(hsv)
ax.scatter(h.flatten(), s.flatten(), v.flatten(), facecolors=pixel_colors, marker=".")

ax.set_xlabel('$h$')
ax.set_ylabel('$s$')
ax.set_zlabel('$v$')
```

```python
rows, cols, _ = img.shape
M = np.float32([[1,0,100],[0,1,50]])

img_predicted = cv.warpAffine(img, M, (cols,rows))
```

```python
from matplotlib.colors import hsv_to_rgb

upper_brown = np.array([30, 255, 155], dtype=np.uint8)
lower_brown = np.array([1,1,1], dtype=np.uint8)

mask = cv2.inRange(hsv, lower_brown, upper_brown)
# mask = cv2.bitwise_not(mask)

```

```python
# mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, np.ones((3,3),np.uint8))
# final_mask = cv2.bitwise_not(mask)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.rcParams["figure.figsize"] = (15,8)

res1 = cv2.bitwise_and(img_rgb, img_rgb, mask=mask)

res2 = cv2.cvtColor(res1, cv2.COLOR_BGR2GRAY)
_, thresh = cv2.threshold(res2, 10, 255, cv2.THRESH_BINARY)

# cv2.imwrite("test.jpg", res1)

plt.imshow(thresh)
plt.title('my picture')
plt.show()

plt.imshow(hsv)
plt.title('hsv')
plt.show()
```
