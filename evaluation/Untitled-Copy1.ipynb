```python
from matplotlib import pyplot as plt
import cv2

path = "12-17-C06F07-B-DBDCDC-B.ihc.HB.jpg"

img = cv2.imread(path)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title('my picture')
plt.show()
```

```python
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img_rgb],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

```python
imghsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

plt.imshow(imghsv)
plt.title('my picture')
plt.show()
```

```python
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([imghsv],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```
