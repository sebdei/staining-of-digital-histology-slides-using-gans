```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

path = "evaluation/testB/22-14-0ECDDB-B-A719A8-B.ihc.HB.jpg"

img = cv2.imread(path )
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

h, s, v = img_hsv[:, :, 0], img_hsv[:, :, 1], img_hsv[:, :, 2]

fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(12,8), dpi= 100)

ax1.set_title("h")
ax1.imshow(h)
ax2.set_title("s")
ax2.imshow(s)
ax3.set_title("v")
ax3.imshow(v)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(12,8), dpi= 100)

ax1.set_title("h")
ax1.hist(h.flatten())
ax2.set_title("s")
ax2.hist(s.flatten())
ax3.set_title("v")
ax3.hist(v.flatten())

plt.show()
```
