```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

path = "evaluation/testB/22-14-0ECDDB-B-A719A8-B.ihc.HB.jpg"

img = cv2.imread(path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

fig, ax1 = plt.subplots(1, 1, figsize=(12,8), dpi=100)


ax1.imshow(img_rgb)
plt.show()
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

h, s, v = img_hsv[:, :, 0], img_hsv[:, :, 1], img_hsv[:, :, 2]

fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(12,8), dpi=100)

ax1.set_title("h")
ax1.imshow(h, cmap = "gray")
ax2.set_title("s")
ax2.imshow(s, cmap = "gray")
ax3.set_title("v")
ax3.imshow(v, cmap = "gray")

plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 4), dpi=100)

ax1.set_title("h")
ax1.hist(h.flatten())
ax2.set_title("s")
ax2.hist(s.flatten())
ax3.set_title("v")
ax3.hist(v.flatten())

plt.show()
```

```python
img_lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

l, a, b = img_lab[:, :, 0], img_lab[:, :, 1], img_lab[:, :, 2]

fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(12,8), dpi=100)

ax1.set_title("l")
ax1.imshow(l, cmap = "gray")
ax2.set_title("a")
ax2.imshow(a, cmap = "gray")
ax3.set_title("b")
ax3.imshow(b, cmap = "gray")

plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 4), dpi=100)

ax1.set_title("l")
ax1.hist(l.flatten())
ax2.set_title("a")
ax2.hist(a.flatten())
ax3.set_title("b")
ax3.hist(b.flatten())

plt.show()
```
